```python
df = pd.read_csv('insurance new.csv',
                #index_col = 'age',
                usecols=['Age','BMI', 'Children', 'Charges', 'Smoker_1', 'Sex_1'])

plt.figure(figsize=(8,12))
heatmap = sns.heatmap(df.corr()[['Charges']].sort_values(by='Charges', ascending = False), 
                      vmin = -1, vmax = 1, annot = True, cmap = 'Blues')
heatmap.set_title('Features Corresponding to Charges', fontdict=
                 {'fontsize':18},pad = 16);
```
